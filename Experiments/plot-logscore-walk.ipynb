## True Function vs Amount of Data

```python
df = df.rename(columns={'test_problem': 'problem'})
```

```python
df = df.replace({"problem": {'function-linear':'linear','function-per':'periodic','function-linplusper':'lin + per','function-linxper':'lin x per'}})
```

```python
problem_order =['linear','periodic','lin + per','lin x per']
```

```python
df = df.replace({"model": {'venture-gp-lin':'Venture GP LIN','venture-gp-se':'Venture GP SE','venture-gp-per':'Venture GP PER','venture-gp-lin-p-per':'Venture GP LIN+PER','venture-gp-lin-t-per':'Venture GP LINxPER','venture_cov_simple':'Venture GP Cov Learning 1','venture_cov_simple_selinper':'Venture GP Cov Learning 2','venture-cov-learning':'Venture GP Cov Learning 3'}})
```

```python
model_dict = {'venture-gp-lin':'Venture GP LIN','venture-gp-se':'Venture GP SE','venture-gp-per':'Venture GP PER','venture-gp-lin-p-per':'Venture GP LIN+PER','venture-gp-lin-t-per':'Venture GP LINxPER','venture_cov_simple':'Venture GP Cov Learning 1','venture_cov_simple_selinper':'Venture GP Cov Learning 2','venture-cov-learning':'Venture GP Cov Learning 3'}
order = ['venture-gp-lin','venture-gp-se','venture-gp-per','venture-gp-lin-p-per','venture-gp-lin-t-per','venture_cov_simple','venture_cov_simple_selinper','venture-cov-learning']
for i in range(len(order)):
    order[i]=model_dict[order[i]]
```

```python
# experiments from the 23. are noise free, so setting it below does not make a difference here
noise_value='0.1'
noise_order=['0.1','0.7']
```

```python
sns.set(font_scale=2) 
df_temp = df.loc[df['n'] == '30']
g = sns.FacetGrid(df_temp.loc[df_temp['noise'] == noise_value], row="problem",margin_titles=False,size=3, aspect=3,hue="model",hue_order=order,palette="coolwarm",row_order=problem_order)
g.map(pl.plot,"logscore");
g.add_legend();
g.set_axis_labels("MCMC Steps", "Log-likelihood");
g.set(ylim=(-200, 0));
g.set(xlim=(0,2000));
```
