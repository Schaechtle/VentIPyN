```python
import seaborn as sns
import pylab as pl
from plotting import load_experiments
import numpy as np
```

```python
sns.set(font_scale=2) 
```

```python
%pylab inline
```

```python
experiment_ini_file ="experiment_stub.ini"
```

```python
date_experiment = "2015-05-19"
```

```python
experimental_df = load_experiments(experiment_ini_file,date_experiment)
```

## True Function vs Amount of Data

```python
order=[str(item) for item in experimental_df['n'].unique()]
```

```python

g = sns.FacetGrid(experimental_df.loc[experimental_df['n'] == '200'], row="test_problem", col="noise", margin_titles=True,size=8, aspect=1,hue="model")
g.map(pl.plot,"logscore");
g.add_legend();
g.set_axis_labels("MCMC Steps", "Log-likelihood");
g.set(ylim=(-1000, 0));
```

```python

g = sns.FacetGrid(experimental_df.loc[experimental_df['test_problem'] == 'function-per'], row="n", col="noise", margin_titles=True,size=8, aspect=1,hue="model",row_order=order)
g.map(pl.plot,"logscore");
g.add_legend();
g.set_axis_labels("MCMC Steps", "Log-likelihood");
g.set(ylim=(-1000, 0));
```

```python
g = sns.FacetGrid(experimental_df.loc[experimental_df['test_problem'] == 'function-linear'], row="n", col="noise", margin_titles=True,size=8, aspect=1,hue="model",row_order=order)
g.map(pl.plot,"logscore");
g.add_legend();
g.set_axis_labels("MCMC Steps", "Log-likelihood");
g.set(ylim=(-1000, 0));
```

```python
g = sns.FacetGrid(experimental_df.loc[experimental_df['n'] == '200'], row="test_problem", col="noise", margin_titles=True,size=8, aspect=1,hue="model")
g.map(pl.errorbar,"index","residuals");
g.add_legend();
g.set_axis_labels("MCMC Steps", "Residuals");
g.set(ylim=(0,40));
```

```python
df_temp=experimental_df.loc[experimental_df['noise'] == '0.1']

g = sns.FacetGrid(experimental_df.loc[experimental_df['noise'] == '0.1'], row="n", col="test_problem", margin_titles=True,size=8, aspect=1,hue="model",row_order=order)
g.map(pl.axhline, y=np.mean(df_temp['base-line']), c="black");
g.map(pl.plot,"residuals");
g.set(ylim=(0,2));
g.add_legend();

```
